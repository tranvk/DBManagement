# Are the pronouns in "Mysterious Japan"  predominantly male or female?


# Open book.txt using file I/O

```python
import string
import re

try:
    with open('japanBook.txt', 'r') as f:
        lines = f.readlines()
        print(lines)

except FileNotFoundError: 
    print("DNE")
    
```

# Clean Text 
1. Remove unwanted characters (\n, punctuation).
    * Use: regex to strip of everything but alphabet/numbers
    
    
2. Store cleaned words in a new list.

3. Change all words to the same case to prevent repeated words (lowercase).



```python
split_words = []

for line in lines:
    words = re.sub("[^a-zA-Z]", " ", line)
    split_words.extend(words.split()) #split by whitespace and store words in list
    
lowercase_words = [words.lower() for words in split_words]
```

# Extract Pronouns Only
* Since we are not interested in the rest of the words, remove everything but the pronouns.
* Function word_frequency: takes list of words as parameter, returns dictionary where (key, value) = (word, count)

```python
copy_list = list(lowercase_words) #keep original list safe
pronouns = ['he', 'she', 'him', 'her']
for word in copy_list:
    if word not in pronouns:
        lowercase_words.remove(word)
        
print(lowercase_words)
```

```python
def word_frequency(word_list):
    frequencies = {}
    for words in word_list:
        if words not in frequencies:
            frequencies[words] = 1
        else:
            frequencies[words] = frequencies[words] + 1
            
    return frequencies
```

```python
pronoun_frequency = word_frequency(lowercase_words)
```

# Consolidate Results
* Add up the pronouns by gender. 

```python
male_count = pronoun_frequency["he"] + pronoun_frequency["him"]
female_count = pronoun_frequency["she"] + pronoun_frequency["her"]
```

```python
print(male_count)
```

```python
print(female_count)
```

# Based on these results, there are predominantly male pronouns in "Mysterious Japan" by Julian Street

```python
ratio = female_count/male_count
```

```python
print(ratio*100)
```
